# Fixed Income Derivatives — Problem Set Week 2

## Problem 1: Interest Rate Swap with Different Payment Frequencies

In this problem, we consider an **interest rate payer swap** where payments on the floating leg occur more frequently than payments on the fixed leg.

Assume the following:
- The fixed leg pays coupons of size $R$, $I$ times per year.
- The floating leg pays **simple LIBOR** coupons **$ J $** times per year.
- The swap matures in exactly **$ N $** years.
- The principal is **$ K $**.
- Then $M = \frac{J}{I} $ and $ N $ are positive integers.

For concreteness, you may think of:
$$
I = 2,\quad J = 4,\quad N = 10.
$$

The swap is considered **at issuance**, at time $ t = T_0 = 0 $.

Let:
- $T_1, T_2, \dots, T_{JN} $ denote the payment dates of the floating leg,
- Fixings are announced at times$ $T_0, T_1, \dots, T_{JN-1}$,
- The LIBOR rate announced at time $T_i$, to be paid at time $T_{i+1}$, is denoted
  $$
  L(T_i, T_{i+1}).
  $$

Assume that **zero-coupon bond prices**$ p(t, T) $ are available for all $T \geq 0 $.


---

### (a) Find an expression for the **value of the fixed leg**.

We know the following:
- Fixed payment dates are: $T_{M}, T_{2M},...,T_{IN*M} = T_{JN}$
- FIxed arrucal period per fixed period: 
$$
\Delta_{i}^{fix} = T_{iM}-T_{(i-1)M}, I=1,...,IN
$$
Thus fixed coupon cashflow at $T_{iM}: RK\Delta_{i}^{fix}$

Discounting each payment using the corresponding zero-coupon bond price \(p(t, T)\), the present value of the fixed leg (ignoring any repayment of principal) is given by
$$
\begin{aligned}
PV_{\text{fixed}}(t)
&= \sum_{i=1}^{IN} p\!\left(t, T_{iM}\right)
\bigl(T_{iM} - T_{(i-1)M}\bigr) R K.
\end{aligned}
$$

It is convenient to define the **accrual factor** (or swap annuity)
$$
S_0
= \sum_{i=1}^{IN} p\!\left(t, T_{iM}\right)
\bigl(T_{iM} - T_{(i-1)M}\bigr),
$$
so that the value of the fixed leg can be written compactly as
$$
\boxed{
PV_{\text{fixed}}(t) = S_0 \, R \, K.
}
$$

---

### (b) Find an expression for the **value of the floating leg**.

The floating leg pays simple LIBOR coupons $J$ times per year at dates  
$T_1, T_2, \dots, T_{JN}$,  
where the fixing announced at time $T_{i-1}$ determines the coupon paid at time $T_i$.  
The coupon paid at $T_i$ is therefore
$$
K \, (T_i - T_{i-1}) \, L(T_{i-1}, T_i).
$$

Discounting each payment using zero-coupon bond prices $p(t,T)$, the present value of the floating leg (ignoring repayment of principal) is
$$
PV_{\text{float}}(t)
= \sum_{i=1}^{JN} p(t,T_i)\,(T_i-T_{i-1})\,L(T_{i-1},T_i)\,K.
$$

Using the standard relation between simple forward LIBOR rates and zero-coupon bond prices,
$$
(T_i-T_{i-1})\,L(T_{i-1},T_i)
= \frac{p(t,T_{i-1})}{p(t,T_i)} - 1,
$$
we obtain
$$
p(t,T_i)\,(T_i-T_{i-1})\,L(T_{i-1},T_i)
= p(t,T_{i-1}) - p(t,T_i).
$$

Hence the floating-leg value telescopes:
$$
\begin{aligned}
PV_{\text{float}}(t)
&= K\sum_{i=1}^{JN}\bigl(p(t,T_{i-1}) - p(t,T_i)\bigr) \\
&= K\bigl(p(t,T_0) - p(t,T_{JN})\bigr).
\end{aligned}
$$

Therefore, the value of the floating leg can be written as
$$
\boxed{
PV_{\text{float}}(t) = K\left(p(t,T_0) - p(t,T_{JN})\right).
}
$$


---

### (c) Par Swap Rate (and where each term comes from)

The **par swap rate** $R_0*$ is the fixed rate that makes the swap have zero value at initiation, i.e.
$$
PV_{\text{float}}(t) - PV_{\text{fixed}}(t) = 0.
$$

Setting $PV_{\text{float}}(t) - PV_{\text{fixed}}(t)=0$ and solving for the par rate gives
$$
K\bigl(p(t,T_0)-p(t,T_{JN})\bigr) - S_0\,R_0^{K} = 0
\quad\Rightarrow\quad
R_0^* = \frac{p(t,T_0)-p(t,T_{JN})}{S_0}.
$$

Equivalently, written with the sum in the denominator:
$$
\boxed{
R_0^*=
\frac{p(t,T_0)-p(t,T_{JN})}
{\sum_{i=1}^{IN} p(t,T_{iM})\,(T_{iM}-T_{(i-1)M})}.
}
$$




---

### (d) Accrual Factor (and where it comes from)

From (c), the par swap rate can be written as
$$
R_0^* = \frac{p(t,T_0)-p(t,T_{JN})}{S_0},
$$
where the **accrual factor** (also called the **swap annuity**) is the denominator:
$$
\boxed{
S_0=\sum_{i=1}^{IN} p(t,T_{iM})\,(T_{iM}-T_{(i-1)M}).
}
$$

It can be interpreted as the **present value of receiving 1 unit of fixed rate per year**, paid according to the fixed-leg schedule. Concretely, each term
$$
p(t,T_{iM})\,(T_{iM}-T_{(i-1)M})
$$
is the discount factor to the fixed payment date $T_{iM}$ times the accrual year fraction for that fixed coupon period.

**Origin of each element**
- $p(t,T_{iM})$: discounting via zero-coupon prices; appears in the **fixed leg** valuation.
- $(T_{iM}-T_{(i-1)M})$: the fixed-leg accrual year fraction; comes from the **fixed leg**.
- The sum $\sum_{i=1}^{IN}(\cdot)$ over fixed payment dates: comes from the **fixed leg** only.

Therefore, the accrual factor $S_0$ stems **solely from the fixed leg**.



### (e) Compare the **par swap rate** and the **accrual factor** you found above to the case where payments to the fixed and floating legs occur at the **same time**.

If payments on the fixed and floating legs occur at the **same dates**, we have $I = J$ and hence $M = J/I = 1$. In that case, the fixed-leg payment dates coincide with the floating-leg payment dates:
$$
T_{iM} = T_i \quad \text{when } M=1.
$$

**Accrual factor (swap annuity)**  
With simultaneous payment dates, the accrual factor becomes
$$
S_0^{\text{sim}}=\sum_{i=1}^{JN} p(t,T_i)\,(T_i-T_{i-1}).
$$
This has the same interpretation as before: the present value of 1 unit paid according to the fixed schedule—now simply the floating schedule because they coincide.

**Par swap rate**  
The floating leg still telescopes to
$$
PV_{\text{float}}(t)=K\bigl(p(t,T_0)-p(t,T_{J*N})\bigr),
$$
and the fixed leg becomes
$$
PV_{\text{fixed}}(t)=K\,R\,S_0^{\text{sim}}.
$$
Setting $PV_{\text{float}}(t)-PV_{\text{fixed}}(t)=0$ yields
$$
R_0^{*,\text{sim}}=\frac{p(t,T_0)-p(t,T_{JN})}{S_0^{\text{sim}}}.
$$

**Conclusion**  
The **structure** of the expressions is the same in both cases:
- Par swap rate is always “floating-leg numerator” divided by the “fixed-leg annuity”.
- The floating-leg numerator $p(t,T_0)-p(t,T_{JN})$ is unchanged.
- Only the **set of dates and accrual periods** used in the accrual factor changes.

Hence, the par swap rate and accrual factor are essentially the same conceptually; the difference is only whether the annuity $S_0$ is built on the coarser fixed schedule ($M>1$) or the common schedule ($M=1$).

---

Now assume that some time has passed and that the current time $ t $ lies **between coupon dates on the floating leg**.

Specifically, assume that slightly more than one year has passed such that:
$$
T_J < t < T_{J+1}.
$$

In this case, the fixing at time $T_J$ has already been announced, and the floating-leg coupon
$$
L(T_J, T_{J+1})
$$
to be paid at time $T_{J+1}$ is **already known**.

---

### (f) Value of the fixed leg at time \(t\)

Fixed-leg payments occur at dates
$$
T_M, T_{2M}, \dots, T_{IN\cdot M},
$$
that is, every \(M\)-th floating-leg payment date. Since \(t > T_J\), the first remaining fixed payment is the first \(T_{iM}\) that satisfies \(T_{iM} > t\). With the indexing used above, this corresponds to
$$
i = \frac{J}{M} + 1.
$$

The present value of the remaining fixed leg (ignoring repayment of principal) is therefore
$$
PV_{\text{fixed}}(t)
=
\sum_{i=\frac{J}{M}+1}^{IN}
p(t,T_{iM})\,(T_{iM}-T_{(i-1)M})\,R\,K.
$$

It is convenient to define the **time-\(t\) accrual factor** (swap annuity)
$$
S_t
=
\sum_{i=\frac{J}{M}+1}^{IN}
p(t,T_{iM})\,(T_{iM}-T_{(i-1)M}),
$$
so that the value of the fixed leg can be written compactly as
$$
\boxed{
PV_{\text{fixed}}(t)=S_t\,R\,K.
}
$$
---

### (g) Find an expression for the **value of the floating leg** at time $ t $.

We assume that time $t$ lies between floating coupon dates such that
$$
T_J < t < T_{J+1}.
$$
At time $t$, the fixing at $T_J$ has already been announced, so the floating coupon to be paid at $T_{J+1}$,
$$
L(T_J, T_{J+1}),
$$
is already known.

The remaining value of the floating leg (ignoring repayment of principal) consists of:
1) the known coupon paid at $T_{J+1}$, plus  
2) the remaining floating payments from $T_{J+2}$ up to $T_{JN}$.

Thus,
$$
PV_{\text{float}}(t)
=
\sum_{i=J+1}^{JN} p(t,T_i)\,(T_i-T_{i-1})\,L(T_{i-1},T_i)\,K.
$$

Separating out the first known cash flow at $T_{J+1}$ gives
$$
\begin{aligned}
PV_{\text{float}}(t)
&= p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})\,K
+ \sum_{i=J+2}^{JN} p(t,T_i)\,(T_i-T_{i-1})\,L(T_{i-1},T_i)\,K.
\end{aligned}
$$

For $i \ge J+2$, we can use the telescoping identity
$$
p(t,T_i)\,(T_i-T_{i-1})\,L(T_{i-1},T_i)
= p(t,T_{i-1}) - p(t,T_i),
$$
so the remaining sum becomes
$$
\sum_{i=J+2}^{JN} K\bigl(p(t,T_{i-1})-p(t,T_i)\bigr)
= K\bigl(p(t,T_{J+1})-p(t,T_{JN})\bigr).
$$

Therefore, the floating-leg value at time $t$ is
$$
\boxed{
PV_{\text{float}}(t)
=
p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})\,K
+
K\bigl(p(t,T_{J+1})-p(t,T_{JN})\bigr).
}
$$


---


### (h) Find expressions for the **par swap rate** and the **accrual factor** at time $ t $.

At time $t$ we have $T_J < t < T_{J+1}$. The par swap rate $R_t^*$ is found by setting the value of the payer swap to zero, i.e.
$$
PV_{\text{float}}(t) - PV_{\text{fixed}}(t) = 0,
$$
and solving for the fixed rate.

From (f), the fixed leg can be written as
$$
PV_{\text{fixed}}(t)=S_t\,R\,K,
$$
where the **accrual factor** (swap annuity) at time $t$ is
$$
\boxed{
S_t
=
\sum_{i=\frac{J}{M}+1}^{IN}
p(t,T_{iM})\,(T_{iM}-T_{(i-1)M}).
}
$$

From (g), the floating leg is
$$
PV_{\text{float}}(t)
=
p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})\,K
+
K\bigl(p(t,T_{J+1})-p(t,T_{JN})\bigr).
$$

Setting $PV_{\text{float}}(t) - PV_{\text{fixed}}(t)=0$ and dividing by $K$ yields
$$
p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})
+
p(t,T_{J+1})-p(t,T_{JN})
-
S_t\,R_t^* = 0.
$$

Hence, the **par swap rate at time $t$** is
$$
\boxed{
R_t^*
=
\frac{
p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})
+
p(t,T_{J+1})-p(t,T_{JN})
}{
\sum_{i=\frac{J}{M}+1}^{IN}
p(t,T_{iM})\,(T_{iM}-T_{(i-1)M})
}.
}
$$

Equivalently, using $S_t$:
$$
\boxed{
R_t^*
=
\frac{
p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})
+
p(t,T_{J+1})-p(t,T_{JN})
}{S_t}.
}
$$

---


### (i) Find an expression for the **PnL of the payer swap**.  Does the expression for the PnL depend on the already fixed rate $ L(T_J, T_{J+1}) $?

Let $R_0^*$ denote the par swap rate at issuance (time $t=0$) and let $R_t^*$ denote the par swap rate at time $t$ (with $T_J < t < T_{J+1}$). The payer swap has fixed rate $R_0^*$.

The PnL at time $t$ can be written as the current value of the swap (since the value at initiation was zero):
$$
PnL(t)=PV_{\text{float}}(t)-PV_{\text{fixed}}(t)\big|_{R=R_0^*}.
$$

Using (f), the fixed leg at time $t$ is
$$
PV_{\text{fixed}}(t)=S_t\,R_0^*\,K,
$$
where
$$
S_t=\sum_{i=\frac{J}{M}+1}^{IN} p(t,T_{iM})\,(T_{iM}-T_{(i-1)M}).
$$

Using (g), the floating leg at time $t$ is
$$
PV_{\text{float}}(t)
=
p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})\,K
+
K\bigl(p(t,T_{J+1})-p(t,T_{JN})\bigr).
$$

Therefore,
$$
\begin{aligned}
PnL(t)
&= p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})\,K
+K\bigl(p(t,T_{J+1})-p(t,T_{JN})\bigr)
- S_t\,R_0^*\,K.
\end{aligned}
$$

Now note from (h) that the par swap rate at time $t$ satisfies
$$
S_t\,R_t^*=
p(t,T_{J+1})\,(T_{J+1}-T_J)\,L(T_J,T_{J+1})
+p(t,T_{J+1})-p(t,T_{JN}).
$$

Substituting this into the PnL expression yields the compact form:
$$
\boxed{
PnL(t)=S_t\,K\,(R_t^*-R_0^*).
}
$$

**Does $PnL(t)$ depend on the already fixed rate $L(T_J,T_{J+1})$?**

In the compact representation above, $PnL(t)$ depends only on:
- the principal $K$,
- the accrual factor $S_t$,
- the change in the par swap rate $(R_t^*-R_0^*)$.

Therefore, the PnL does **not** depend on the already fixed floating rate $L(T_J,T_{J+1})$ explicitly once expressed in terms of par swap rates and the annuity:
$$
\boxed{\text{No, }PnL(t)\text{ does not depend on }L(T_J,T_{J+1})\text{ in this form.}}
$$

# Fixed Income Derivatives — Problem Set Week 2  
## Problem 2: Term Structure Construction and Applications

For this problem, assume for simplicity that:
- A year consists of **12 months**, each with exactly **30 days**,
- All payments occur at the **end of the last day of the month**,
- There is **no credit risk**,
- The principal of all bonds is **100**.

Today’s date is **December 30, 2019**, the last day of the year.  
At **11:00 AM**, the BBA has announced the following LIBOR fixings:
- 3M LIBOR: $0.01570161$,
- 6M LIBOR: $0.01980204$.

In addition, the following bonds are traded in the market:

1. A **3-year** fixed-rate bullet bond issued on December 30, 2015, maturing December 30, 2020, paying **quarterly simple coupons** of **4% annually**, trading at **102.33689177**.
2. A **5-year** fixed-rate bullet bond issued on December 30, 2010, maturing December 30, 2020, paying **semi-annual simple coupons** of **5% annually**, trading at **104.80430234**.
3. A **10-year** fixed-rate bullet bond issued on December 30, 2011, maturing June 30, 2021, paying **semi-annual simple coupons** of **5% annually**, trading at **105.1615306**.
4. An **8-year** fixed-rate bullet bond issued on December 30, 2016, maturing June 30, 2021, paying **quarterly simple coupons** of **6% annually**, trading at **105.6581905**.
5. A **5-year** fixed-rate bullet bond issued on December 30, 2011, maturing December 30, 2021, paying **quarterly simple coupons** of **5% annually**, trading at **104.028999992**.
6. A **30-year** fixed-rate bullet bond issued on December 30, 2001, maturing December 30, 2021, paying **annual simple coupons** of **3% annually**, trading at **101.82604116**.

---



### (a)
Set up the **cash-flow matrix** corresponding to the above information.

---


```python
import numpy as np
import pandas as pd

# Time grid (years from today, 30/360 convention)
T = np.array([0, 0.25, 0.50, 0.75, 1.00, 1.25, 1.50, 1.75, 2.00])

# Initialize cash-flow matrix: 6 bonds × 9 dates
C = np.zeros((6, len(T)))

# (i) 3Y bond, quarterly 4%, matures at 1.00
C[0, 0:4] = 1.0
C[0, 4]   = 101.0

# (ii) 5Y bond, semi-annual 5%, matures at 1.00
C[1, 0] = 2.5
C[1, 2] = 2.5
C[1, 4] = 102.5

# (iii) 10Y bond, semi-annual 5%, matures at 1.50
C[2, 0] = 2.5
C[2, 2] = 2.5
C[2, 4] = 2.5
C[2, 6] = 102.5

# (iv) 8Y bond, quarterly 6%, matures at 1.50
C[3, 0:6] = 1.5
C[3, 6]   = 101.5

# (v) 5Y bond, quarterly 5%, matures at 2.00
C[4, :] = 1.25
C[4, 8] = 101.25

# (vi) 30Y bond, annual 3%, matures at 2.00
C[5, 0] = 3.0
C[5, 4] = 3.0
C[5, 8] = 103.0

# Prices
prices = np.array([
    102.33689177,
    104.80430234,
    105.1615306,
    105.6581905,
    104.028999992,
    101.82604116
])

# Display nicely
df_C = pd.DataFrame(C, columns=T,
                    index=["Bond (i)", "Bond (ii)", "Bond (iii)",
                           "Bond (iv)", "Bond (v)", "Bond (vi)"])

df_C, prices
```


### (b)
Find the vector of **zero-coupon prices** for all maturities that can be identified from the data, and compute the **continuously compounded zero-coupon spot rates** (the yield curve).  
Report the results and plot both curves in an appropriate diagram.

---


We can immediately identify the shortest discount factors:
- $p(0,0)=1$,
- $p(0,0.25)=\frac{1}{1+L_{3M}\cdot 0.25}$ from 3M LIBOR,
- $p(0,0.50)=\frac{1}{1+L_{6M}\cdot 0.50}$ from 6M LIBOR.

Bond prices satisfy the linear pricing system
$$
\pi = C\,p,
$$
where $C$ is the cash-flow matrix and $p$ is the vector of zero-coupon bond prices on the time grid.
Since $C$ is $6\times 9$, the system is underdetermined. However, because $p(0,0.25)$ and $p(0,0.50)$ are known from LIBOR, we can move their contributions to the left-hand side and solve the remaining **square** $6\times 6$ system for the remaining ZCB prices.


```python
import numpy as np
import pandas as pd

L3M = 0.01570161
L6M = 0.01980204

# Indices in the grid
idx_0   = int(np.where(np.isclose(T, 0.00))[0][0])
idx_025 = int(np.where(np.isclose(T, 0.25))[0][0])
idx_050 = int(np.where(np.isclose(T, 0.50))[0][0])

# Build augmented matrix A (9x9) and RHS b (9,)
A = np.zeros((9, 9))
b = np.zeros(9)

# 1) Bond equations: C p = prices
A[0:6, :] = C
b[0:6] = prices

# 2) p(0,0) = 1
A[6, idx_0] = 1.0
b[6] = 1.0

# 3) (1 + L3M*0.25) p(0,0.25) = 1
A[7, idx_025] = 1.0 + L3M * 0.25
b[7] = 1.0

# 4) (1 + L6M*0.50) p(0,0.50) = 1
A[8, idx_050] = 1.0 + L6M * 0.50
b[8] = 1.0

# Solve for all ZCB prices on the grid
p = np.linalg.solve(A, b)

# Continuous spot rates
R = np.full_like(T, np.nan, dtype=float)
mask = T > 0
R[mask] = -np.log(p[mask]) / T[mask]

df = pd.DataFrame({"T": T, "p(0,T)": p, "R(0,T) cc": R})
df
```

```python
import matplotlib.pyplot as plt

# We can then plot the zero-coupon prices and the continuously compounded spot rates, in the same diagram.
fig, ax1 = plt.subplots(figsize=(10, 5))

color1 = 'tab:blue'
ax1.set_xlabel('T (years)')
ax1.set_ylabel('ZCB Price p(0,T)', color=color1)
l1 = ax1.plot(T, p, color=color1, marker='o', label='ZCB Price p(0,T)')
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Spot Rate R(0,T) cc', color=color2)
l2 = ax2.plot(T, R, color=color2, marker='s', label='Spot Rate R(0,T) cc')
ax2.tick_params(axis='y', labelcolor=color2)

lns = l1 + l2
labels = [l.get_label() for l in lns]
ax1.legend(lns, labels, loc="best")

plt.title("Zero-Coupon Bond Prices and Spot Rates")
plt.show()



```


### (c)
Find the **3M forward rates** and plot them in the diagram from part (b).

Given the zero-coupon prices $p(0,T)$ from part (b), the **3M simple forward rate** for the period $[T, T+0.25]$ is defined by
$$
1 + 0.25\,F(0;T,T+0.25) = \frac{p(0,T)}{p(0,T+0.25)}.
$$

Hence,
$$
\boxed{
F(0;T,T+0.25)=\frac{1}{0.25}\left(\frac{p(0,T)}{p(0,T+0.25)}-1\right).
}
$$

We compute these for $T \in \{0,0.25,0.50,\dots,1.75\}$ and plot them together with the curves from part (b).

---


```python
# 1) Find the 3M forward rates
dt = 0.25  # 3 months

# Forward rates on each 3M interval in the grid
F3M = (p[:-1] / p[1:] - 1.0) / dt
T_fwd = T[:-1]  # forward rate for interval [T_i, T_{i+1}]

# Plot: ZCB prices and spot rates from (b) + forward rates from (c)
fig, ax1 = plt.subplots()

# ZCB prices (left axis)
ax1.plot(T, p, marker="o")
ax1.set_xlabel("Maturity T (years)")
ax1.set_ylabel("ZCB price p(0,T)")

# Spot rates + forward rates (right axis)
ax2 = ax1.twinx()
ax2.plot(T[1:], R[1:], marker="o")        # spot rates (skip T=0)
ax2.plot(T_fwd, F3M, marker="o")          # 3M forward rates
ax2.set_ylabel("Rates (continuous spot and 3M forward)")

plt.title("ZCB prices, spot rates, and 3M forward rates")

ax1.legend(['ZCB price'], loc="upper left")
spot_color = 'purple'  # New color not already used (not blue, orange, etc.)
forward_color = 'green'  # Optional: distinguish forward rates as well

l1, = ax2.plot(T[1:], R[1:], marker="o", color=spot_color, label='Spot rate')        # spot rates (skip T=0)
l2, = ax2.plot(T_fwd, F3M, marker="o", color=forward_color, label='3M forward rate') # 3M forward rates
ax2.legend(handles=[l1, l2], loc="upper right")




plt.show()
```


### (d)
Find the price of a **2-year floating-rate bullet note** with principal 100 paying **6M LIBOR**, issued today.

---
Consider a floating-rate note (FRN) with principal $K=100$ that pays **6M LIBOR** every 6 months over 2 years. Let the payment dates be
$$
0.5,\;1.0,\;1.5,\;2.0.
$$
Under no-arbitrage, the value of a par floater (ignoring credit risk) at issuance is equal to the principal, since the floating coupons reset to the market and replicate rolling short discount bonds. Equivalently, using the telescoping argument,
$$

### (e)
Find the **par swap rate** for a **2-year interest rate swap** paying:
- semi-annual fixed coupons at annual rate $R$ to the receiver,
- quarterly 3M LIBOR to the payer.

---
Using the standard swap decomposition at initiation, the PV of the floating leg (ignoring principal exchange) is
$$
PV_{\text{float}}(0)=K\bigl(p(0,0)-p(0,2)\bigr)=K\bigl(1-p(0,2)\bigr).
$$

The PV of the fixed leg is
$$
PV_{\text{fixed}}(0)=K\,R\,A,
\qquad
A=\sum_{i=1}^{4} 0.5\,p(0,0.5i).
$$

The **par swap rate** $R^*$ is found by setting $PV_{\text{float}}(0)-PV_{\text{fixed}}(0)=0$, giving
$$
\boxed{
R^*=\frac{1-p(0,2)}{\sum_{i=1}^{4} 0.5\,p(0,0.5i)}.
}
$$

```python
K = 100.0

# Fixed leg dates (semi-annual) within 2 years
T_fix = np.array([0.5, 1.0, 1.5, 2.0])
dt_fix = 0.5

# Get indices for these maturities in the grid T
idx_fix = [int(np.where(np.isclose(T, Ti))[0][0]) for Ti in T_fix]

# Fixed-leg annuity
A = np.sum(dt_fix * p[idx_fix])

# Par swap rate
R_par = (1.0 - p[int(np.where(np.isclose(T, 2.0))[0][0])]) / A

print("Fixed-leg annuity A:", A)
print("Par swap rate R*:", R_par)
```


### (f)
Compare the par swap rate found in part (e) to the forward rates computed in part (c).

From (c), the 3M simple forward rate for the interval $[T,T+0.25]$ is
$$
F(0;T,T+0.25)=\frac{1}{0.25}\left(\frac{p(0,T)}{p(0,T+0.25)}-1\right).
$$

The par swap rate $R^*$ is **not equal to a single 3M forward rate**. Instead, it can be interpreted as an **annuity-weighted average** of the expected short rates over the swap’s life (under the relevant pricing measure), because:
- the floating leg aggregates many quarterly forward rates,
- while the fixed leg is an annuity (sum of discounted accrual periods).

Hence, $R^*$ typically lies “in the same region” as the sequence of forward rates over $[0,2]$, but it is a **weighted average** determined by the discount factors $p(0,T)$ via the fixed-leg annuity.
